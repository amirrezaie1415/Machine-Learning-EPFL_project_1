The histograms show that there are some data that are not consistenet with the remaining data and they are all equal to **`-999`**. In the code below, we count the number of samples having at least one feature value equal to -999.


```python
invalid_indices = set()     # indices of samples having at least one feature equal to -999" 
for featureID in range(X.shape[1]):
    index = np.where(X[:,featureID] == -999.0)[0]
    invalid_indices.update(index.tolist())
```

```python
print('Number of samples having at least one feature equal to -999:', len(invalid_indices))
print('Ratio of the number of samples to be deleted to total number of samples:', round(len(invalid_indices)/len(X),2))
```

From the code above, it can be seen that if we delete the samples having at least one invalide feature value, we need to delete almost $73 \%$ of our samples. 

# Outliers

```python
# removing samples having at least one outlier in their feature vectors
outlier_indices = [202795, 191362, 182617, 173649, 171360, 153775, 151832, 150377, 132402, 118381, 117359,
 84175, 75797, 68117,  52754, 49297, 19843, 7343]
X = np.delete(X, outlier_indices, axis = 0) 
y = np.delete(y, outlier_indices, axis = 0)
ids = np.delete(ids, outlier_indices, axis = 0)
```

```python
labels = ['0', '1', '2', '3']
category_count =np.array( [count_categorical(y,X,0)[0] + count_categorical(y,X,0)[1],
                  count_categorical(y,X,1)[0] + count_categorical(y,X,1)[1],
                  count_categorical(y,X,2)[0] + count_categorical(y,X,2)[1],
                  count_categorical(y,X,3)[0] + count_categorical(y,X,3)[1]])

x = np.arange(len(labels))
fig, ax = plt.subplots()
rects1 = ax.bar(x, category_count, 0.2, label='signal/background', color=['blue', 'blue', 'blue', 'blue'])
ax.set_ylabel('Count')
ax.set_xlabel('PRI_jet_num')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```
