## 1- Toy example 1: 
In this example, i.e. `toy example 1`, the implemented functions `least_squares`, `least_squares_GD`, `least_squares_SGD` and `ridge_regression` are verified. 

Given a **linear function** defined as $\left \{ f_{\mathbf{w}}(x) = w_{0} + w_{1} x| \hspace {0.1cm}(f_{\mathbf{w}}(x), x) \in \mathbb{R} \times \mathbb{R} \right \}$, where $w_{0} = 5$ ,$w_{1} = 10$,

a **training set** $(x^{(1)}, y^{(1)}), ...,(x^{(n)},
y^{(n)}) \in \mathbb{R} \times \mathbb{R}$ is created, where $x^{(n)} \in \left [ 0,10 \right )$ with a step size of $0.1$.

```python
# Linear model of the form y= 10*x+5 (w0 = 5.0 w_1 = 10.0)
w_0, w_1 = 5.0, 10.0
x = np.arange(start=0, stop=10.0, step=0.1)
y = w_1 * x + w_0
plt.rcParams.update({'font.size': 14})
plt.plot(x, y, 'o', c='r', markersize=2)
plt.xlabel('x')
plt.ylabel('y')
plt.text(w_0+1,w_1+1,'f(x) = {w1} x + {w0}'.format(w1=w_1, w0=w_0))
plt.title('Training datapoints generated from a linear model.')
plt.show()
```

By assimilating the intercept $w_{0}$ into $\mathbf{w}$:

 * Adding a new feature that is 1: $\tilde{x} = (1,x) \in \mathbb{R}^{2}$
 * Setting $\tilde{w} = (w_{0}, w_{1})$ 
 
Then $f_{\mathbf{\tilde{w}}}(x) = \mathbf{\tilde{w}}.\mathbf{\tilde{x}}$

```python
# add a column of 1 to the desing matrix
x_tilda = np.c_[np.ones(x.shape[0]),x]
```

```python
y = y.reshape(y.shape[0],1)
```

```python
y.shape, x_tilda.shape 
```

## 1-1 Solution of the implemented functions 

### 1-1-1 Solution of the implemented function "least_squares" 

```python
# test the function "least_squares"
w_least_squares, loss_least_squares = least_squares(y, x_tilda)
```

```python
print('weight and loss values obtianed from the "least_squares" function:')

print('w0 = {w0:.2f}, w1={w1:.2f}, loss = {l:.2e}'.format(w0 = w_least_squares[0,0], w1 = w_least_squares[1,0], l = loss_least_squares))
```

### 1-1-2 Solution of the implemented function "least_squares_GD" 

```python
# test the function "least_squares_GD"
initial_w = np.array([[1.0], [1.0]], dtype = float)
gamma = 0.01
max_iters = 2000
w_least_squares_GD, loss_least_squares_GD = least_squares_GD(y, x_tilda, initial_w, max_iters, gamma)
```

```python
print('weight and loss values obtianed from the "least_squares_GD" function:')
print('w0 = {w0:.2f}, w1={w1:.2f}, loss = {l:.2e}'.format(
    w0 = w_least_squares_GD[0,0], w1 = w_least_squares_GD[1,0], l = loss_least_squares_GD))
```

### 1-1-3 Solution of the implemented function "least_squares_SGD" 

```python
# test the function "least_squares_SGD"
initial_w = np.array([[1.0], [1.0]], dtype = float)
gamma = 0.01
max_iters = 80
w_least_squares_SGD, loss_least_squares_SGD = least_squares_SGD(y, x_tilda, initial_w, max_iters, gamma)
```

```python
print('weight and loss values obtianed from the "least_squares_SGD" function:')
print('w0 = {w0:.2f}, w1={w1:.2f}, loss = {l:.2e}'.format(
    w0 = w_least_squares_SGD[0,0], w1 = w_least_squares_SGD[1,0], l = loss_least_squares_SGD))
```

### 1-1-4 Solution of the implemented function "ridge_regression" 

```python
# test the function "ridge_regression"
lambda_ = 0.1
w_ridge_regression, loss_ridge_regression = ridge_regression(y, x_tilda, lambda_)
```

```python
print('weight and loss values obtianed from the "ridge_regression" function:')
print('w0 = {w0:.2f}, w1={w1:.2f}, loss = {l:.2e}'.format(
    w0 = w_ridge_regression[0,0], w1 = w_ridge_regression[1,0], l = loss_ridge_regression))
```

## 1-2 Comparing results

Now the estimated $w_{0}, w_{1}$ from the implemented functions `"least_squares"`, `"least_squares_GD"` and `least_squares_SGD` are compared with the true values. It can be seen that the functions can accurately predict the vector $\mathbf{w}$.

```python
print('true values \t least_squares function \t least_squares_GD function  \t least_squares_SGD function ' )
print('w0\t w1 \t w0\tw1 \t\t\t w0\tw1 \t\t\t w0\tw1')
print('{:.3}\t{:.3}\t{:.3}\t{:.3}\t\t\t{:.3}\t{:.3}\t\t\t{:.3}\t{:.3}'.format(
    w_0,w_1,w_least_squares[0,0], w_least_squares[1,0], w_least_squares_GD[0,0], w_least_squares_GD[1,0], w_least_squares_SGD[0,0], w_least_squares_SGD[1,0]))
```

## 2- Toy example 2: 

In this example, i.e. `toy example 2` **[2]**, the implemented functions `logistic_regression`, and `reg_logistic_regression` are verified.

```python
# Generate a toy dataset 
xmin, xmax = -5, 5
n_samples = 100
np.random.seed(0)
X = np.random.normal(size=n_samples)
y = (X > 0).astype(np.float)
X[X > 0] *= 4
X[ X < 0] -= 5
X = X[:, np.newaxis]
X_tilda = np.c_[np.ones(X.shape[0]),X]
y = y.reshape(y.shape[0],1)
```

```python
# plot the trainig data
plt.rcParams.update({'font.size': 14})
plt.scatter(X.ravel(), y, color='blue', zorder=20)
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Training data of the "toy example 2"')
plt.show()
```

## 2-1 Solution of the implemented functions 


### 2-1-1 Solution of the implemented function "logistic_regression" 

```python
# fitting the logistic regression function
gamma = 0.05
max_iters = 500
initial_w = np.ones((X_tilda.shape[1],1), dtype=float)
w_logist, loss_logist = logistic_regression(y, X_tilda, initial_w, max_iters, gamma)
```

```python
# plotting the fitted function
X_test_tilda = np.linspace(-9, 10, 300)
X_test_tilda = np.c_[np.ones(X_test_tilda.shape[0]),X_test_tilda]
y_test_without_regularizer = logistic_function(X_test_tilda @ w_logist)
plt.rcParams.update({'font.size': 14})
plt.scatter(X.ravel(), y, color='blue', s=50, marker='o', facecolor='none' ,label='Training data')
plt.plot(X_test_tilda[:,1], y_test_without_regularizer, color='red', label='Fitted function')
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Fitting logistic regression function to the "toy example 2 data"')
plt.legend(loc=(1.04,0.7))
plt.show()
```

### 2-1-2 Solution of the implemented function "reg_logistic_regression" 

```python
# fitting the logistic regression function
gamma = 0.05
max_iters = 500
lambda_ = 0.1
initial_w = np.ones((X_tilda.shape[1],1), dtype=float)
w_reg_logist, loss_reg_logist = reg_logistic_regression(y, X_tilda,lambda_ ,initial_w, max_iters, gamma)
```

```python
# plotting the fitted function
plt.rcParams.update({'font.size': 14})
X_test_tilda = np.linspace(-9, 10, 300)
X_test_tilda = np.c_[np.ones(X_test_tilda.shape[0]),X_test_tilda]
y_test_with_regularizer = logistic_function(X_test_tilda @ w_reg_logist)
plt.scatter(X.ravel(), y, color='blue', s=50, marker='o', facecolor='none' ,label='Training data')
plt.plot(X_test_tilda[:,1], y_test_with_regularizer, color='red', label='Fitted function')
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Fitting regularized logistic regression function to the "toy example 2 data"')
plt.legend(loc=(1.04,0.7))
plt.show()
```

### 2-1-2 Solution of the implemented function "reg_logistic_regression" 

```python
# fitting the logistic regression function
gamma = 0.05
max_iters = 500
lambda_ = 10
initial_w = np.ones((X_tilda.shape[1],1), dtype=float)
w_reg_logist, loss_reg_logist = reg_logistic_regression(y, X_tilda,lambda_ ,initial_w, max_iters, gamma)
```

```python
# plotting the fitted function
plt.rcParams.update({'font.size': 14})
X_test_tilda = np.linspace(-9, 10, 300)
X_test_tilda = np.c_[np.ones(X_test_tilda.shape[0]),X_test_tilda]
y_test_with_regularizer = logistic_function(X_test_tilda @ w_reg_logist)
plt.scatter(X.ravel(), y, color='blue', s=50, marker='o', facecolor='none' ,label='Training data')
plt.plot(X_test_tilda[:,1], y_test_with_regularizer, color='red', label='Fitted function')
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Fitting regularized logistic regression function to the "toy example 2 data"')
plt.legend(loc=(1.04,0.7))
plt.show()
```

## 2-2 Comparing the results

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(8,8))
plt.scatter(X.ravel(), y, color='blue', s=50, marker='o', facecolor='none' ,label='Training data')
plt.plot(X_test_tilda[:,1], y_test_without_regularizer, color='red', label='Fitted function without regularizer')
plt.plot(X_test_tilda[:,1], y_test_with_regularizer, color='green', label='Fitted function with regularizer')
plt.xlabel('x')
plt.ylabel('probability')
plt.title('Fitting regularized logistic regression function to the "toy example 2 data"')
plt.legend(loc=(1.04,0.7))
plt.show()
```
